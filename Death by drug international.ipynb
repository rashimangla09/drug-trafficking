```python
data = pd.read_csv("deaths data")
```

```python
# Convert 'Deaths that are from all causes attributed to drug use' column to numeric
data['Deaths that are from all causes attributed to drug use, in both sexes aged all ages'] = pd.to_numeric(data['Deaths that are from all causes attributed to drug use, in both sexes aged all ages'], errors='coerce')
```

```python
#visualization
```

```python
# Group data by country and sum the deaths
death_by_country = data.groupby('COUNTRY')['Deaths - Drug use disorders - Sex: Both - Age: All Ages (Number)'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=death_by_country.values, y=death_by_country.index, palette='viridis')
plt.xlabel('Total Deaths')
plt.ylabel('Country')
plt.title('Top 10 Countries by Total Deaths')
plt.show()
```

```python
pivot_table = data.pivot_table(index='COUNTRY', columns='Year', values='Deaths - Drug use disorders - Sex: Both - Age: All Ages (Number)', aggfunc='sum')
plt.figure(figsize=(12,8))
sns.heatmap(pivot_table, cmap='YlGnBu')
plt.title('Deaths Distribution by Country and Year')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data['Deaths - Drug use disorders - Sex: Both - Age: All Ages (Number)'], data['Deaths that are from all causes attributed to drug use, in both sexes aged all ages'], alpha=0.5)
plt.xlabel('Deaths from Drug Use Disorders')
plt.ylabel('Deaths from All Causes Attributed to Drug Use')
plt.title('Comparison of Deaths from Drug Use Disorders vs. All Causes')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='Year', y='Deaths - Drug use disorders - Sex: Both - Age: All Ages (Number)', data=data)
plt.xlabel('Year')
plt.ylabel('Total Deaths')
plt.title('Trend of Deaths Over Time')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(data['Deaths - Drug use disorders - Sex: Both - Age: All Ages (Number)'], bins=20, color='blue')
plt.xlabel('Total Deaths')
plt.ylabel('Frequency')
plt.title('Distribution of Deaths')
plt.show()
```
